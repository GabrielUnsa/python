```
'''
Clase que contiene:
  * Color: Numero Entero
  * Vertices: Que pertenencen al color, se maneja como un conjunto( set )
'''
class color:
  def __init__( self, color = None, setVertex = None, function = None ):
    self.color = color
    self.setVertex = setVertex

  def set_color( self, color ):
    self.color = color
    
  def get_color( self ):
    return self.color

  def set_allVertex( self, vertex ):
    self.setVertex = vertex

  #Devuelve el conjunto de vertices
  def get_allVertex( self ):
    return self.setVertex

  #AÃ±ade un vertice al conjunto
  def addVertex( self, vertex ):
    if not self.setVertex:
      self.setVertex = set()
      self.setVertex.add( vertex )
    else:
      if not ( vertex in self.setVertex ):
        self.setVertex.add( vertex )
  
  #Eliminar un vertice al conjunto
  def remVertex( self, vertex ):
    self.setVertex.remove( vertex )  
  
  #Devuelve True si el Vertice esta ya en el conjunto sino devuelve False
  def isVertex( self, vertex ):
    return( vertex in self.setVertex )
  
  #Muestra el conjunto de vertices
  def printVertex( self ):
    for vertex in self.setVertex:
      print(vertex, end = " => " )
    print()
    
```
