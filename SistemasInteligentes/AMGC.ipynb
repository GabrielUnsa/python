```
#Lista enlazada de la solucion
class nodeSolution:
  def __init__( self, solution = None, next = None):
    self.solution = solution
    self.next = next

class linked_list:
  def __init__( self ):
    self.head = None
  
  def add_front( self, solution ):
    self.head = nodeSolution( solution = solution, next = self.head )
    
  def is_empty( self ):
    return sel.head == None
  
  def add_end( self, solution):
    if not self.head:
      self.head = nodeSolution( solution = solution )
      return
    curr = self.head
    while curr.next:
      curr = curr.next
    curr.next = nodeSolution( solution = solution)
    
  def delete_node( self, key ):
    curr = self.head
    prev = None
    while curr and curr.solution != key:
      prev = curr
      curr = curr.next
    if prev is None:
      self.head = curr.next
    elif curr:
      prev.next = curr.next
      curr.next = None

  def get_last_node(self):
    temp = self.head
    while(temp.next is not None):
        temp = temp.next
    return temp.solution
    
  def print_list( self ):
    node = self.head
    while node != None:
        print(node.solution, end =" => ")
        node = node.next
    print()
```
